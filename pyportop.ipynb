```python
import os
# Parâmetros definidos pelo usuário
target_vol = 0.06       # volatilidade alvo
lambda_pen = 1e3        # penalização de volatilidade
tol = 0.005             # tolerância para as restrições lineares
eps = 1e-6              # epsilon para zero
n_jobs = max(1, os.cpu_count() - 1)
peso_limite_pct = 1  # limite de alteração como percentual da posição original (15% por padrão)
```

```python
import pandas as pd

# --- caminhos para os arquivos Excel ----------------------------
# (mantenha-os como strings simples se o .py estiver na mesma pasta)
ARQ_PRECOS_LONGOS   = "precos_longos.xlsx"
ARQ_POSICOES_NET    = "posicoes_net.xlsx"
ARQ_POSICOES_IBOV   = "posicoes_Ibov.xlsx"
ARQ_POSICOES_SMAL   = "posicoes_Smal.xlsx"

# --- leitura em DataFrames --------------------------------------
df_precos_longos = pd.read_excel(ARQ_PRECOS_LONGOS,   sheet_name=0, index_col=0)
df_posicoes_net  = pd.read_excel(ARQ_POSICOES_NET,    sheet_name=0, index_col=0)
df_posicoes_ibov = pd.read_excel(ARQ_POSICOES_IBOV,   sheet_name=0, index_col=0)
df_posicoes_smal = pd.read_excel(ARQ_POSICOES_SMAL,   sheet_name=0, index_col=0)

# --- Display em DataFrames --------------------------------------
print('df_posicoes_ibov')
display(df_posicoes_ibov)
print('df_posicoes_net')
display(df_posicoes_net)
print('df_posicoes_smal')
display(df_posicoes_smal)
print('df_precos_longos')
display(df_precos_longos)

# --- confirmação rápida (opcional) ------------------------------
print("df_precos_longos:", df_precos_longos.shape)
print("df_posicoes_net :", df_posicoes_net.shape)
print("df_posicoes_ibov:", df_posicoes_ibov.shape)
print("df_posicoes_smal:", df_posicoes_smal.shape)
```

```python
import numpy as np
import pandas as pd

def calculate_expost_volatility(
    df_precos_longos: pd.DataFrame,
    df_posicoes_net: pd.DataFrame,
    df_posicoes_ibov: pd.DataFrame,
    df_posicoes_smal: pd.DataFrame
) -> pd.Series:
    # Renomeia colunas de índices para coincidir com df_precos_longos
    df_ibov = df_posicoes_ibov.rename(columns={'peso_short_ibov': 'IBOV'})
    df_smal = df_posicoes_smal.rename(columns={'peso_short_smal': 'SMAL11'})
    
    # Concatena todos os pesos em um único DataFrame
    df_weights = pd.concat([df_posicoes_net, df_ibov, df_smal], axis=1)
    
    # Seleciona apenas os ativos presentes nos pesos
    assets = df_weights.columns.tolist()
    df_prices = df_precos_longos[assets]
    
    # Calcula retornos simples diários e preenche NaNs com 0
    df_returns = df_prices.pct_change().fillna(0)
    
    # Alinha datas entre retornos e pesos
    common_dates = df_weights.index.intersection(df_returns.index)
    df_weights = df_weights.loc[common_dates]
    
    # Converte para arrays NumPy para otimização
    returns_values = df_returns.values
    dates = df_returns.index
    date_to_loc = {date: idx for idx, date in enumerate(dates)}
    weight_locs = [date_to_loc[date] for date in df_weights.index]
    weight_values = df_weights.values
    
    # Calcula volatilidade ex-post para cada dia de backtest
    vol_list = []
    for loc, w in zip(weight_locs, weight_values):
        start = loc - 252 + 1
        if start < 0:
            raise ValueError(f"Histórico insuficiente para calcular volatilidade no índice {loc}")
        window_returns = returns_values[start:loc+1, :]
        r_p = window_returns.dot(w)
        vol = r_p.std(ddof=1) * np.sqrt(252)
        vol_list.append(vol)
    
    # Retorna Série com volatilidades indexada pelas datas de backtest
    return pd.Series(vol_list, index=df_weights.index, name='vol_expost')

vol_expost = calculate_expost_volatility(df_precos_longos,df_posicoes_net,df_posicoes_ibov,df_posicoes_smal)
print('\n series vol_expost')
print(vol_expost)
```

```python
import os
import numpy as np
import pandas as pd
from scipy.optimize import minimize
from joblib import Parallel, delayed
from tqdm import tqdm
import time



def volatility_seeking_parallel(
    df_precos_longos: pd.DataFrame,
    df_posicoes_net: pd.DataFrame,
    df_posicoes_ibov: pd.DataFrame,
    df_posicoes_smal: pd.DataFrame,
    datas_otimizacao: pd.DataFrame = None
) -> (pd.DataFrame, pd.Series, pd.Series):
    """
    Computa pesos ajustados e volatilidade ex-post buscada em paralelo para cada dia do backtest,
    usando sempre as últimas 252 observações do df_precos_longos (ajustado).
    
    Parâmetros:
      df_precos_longos: DataFrame com preços ajustados
      df_posicoes_net: DataFrame com posições net
      df_posicoes_ibov: DataFrame com posições IBOV
      df_posicoes_smal: DataFrame com posições SMAL11
      datas_otimizacao: DataFrame com datas específicas para otimização

    Retorna:
      df_weights_adj: DataFrame com pesos ajustados por dia
      series_vol_seeking: Series com volatilidade ex-post alcançada por dia
      series_converged: Series booleana de convergência do solver
    """
    # Medir tempo de execução
    tempo_inicio = time.time()
    
    print("Preparando dados para otimização...")
    
    # 1. Preparação básica - mais eficiente com merge em vez de concat
    df_ibov = df_posicoes_ibov.rename(columns={'peso_short_ibov': 'IBOV'})
    df_smal = df_posicoes_smal.rename(columns={'peso_short_smal': 'SMAL11'})
    
    # Usando merge em vez de concat para melhor performance
    df_weights = df_posicoes_net.copy()
    df_weights['IBOV'] = df_ibov['IBOV']
    df_weights['SMAL11'] = df_smal['SMAL11']
    df_weights = df_weights.loc['2019-09-02':]
    
    # Convertemos para numpy arrays para cálculos mais rápidos
    assets = df_weights.columns.tolist()
    n_assets = len(assets)
    ibov_idx = assets.index('IBOV')
    smal_idx = assets.index('SMAL11')
    
    # Retornos sobre toda a série de preços ajustados
    # Seleciona apenas as colunas necessárias
    colunas_necessarias = [col for col in df_weights.columns if col in df_precos_longos.columns]
    df_returns = df_precos_longos[colunas_necessarias].pct_change().fillna(0)
    
    # Datas de backtest - filtrar apenas pelas datas especificadas, se fornecidas
    all_dates = df_weights.index.intersection(df_returns.index)
    
    if datas_otimizacao is not None and not datas_otimizacao.empty:
        try:
            # Carregar as datas de otimização do DataFrame
            datas_otim = pd.to_datetime(datas_otimizacao['Data_otimizacao'])
            # Filtrar apenas as datas que estão em all_dates
            dates = all_dates.intersection(datas_otim)
            print(f"Otimizando apenas para {len(dates)} datas específicas em vez de {len(all_dates)} datas.")
        except Exception as e:
            print(f"Erro ao processar datas de otimização: {e}")
            dates = all_dates
    else:
        dates = all_dates
    
    # 2. Pré-computa todas as covariâncias rolling - fazemos em paralelo para acelerar
    print("Pré-calculando matrizes de covariância...")
    
    def compute_cov(date):
        return date, df_returns.loc[:date].iloc[-252:].cov().values
    
    cov_results = Parallel(n_jobs=n_jobs)(
        delayed(compute_cov)(d) for d in tqdm(dates, desc="Calculando covariâncias")
    )
    cov_dict = {date: cov for date, cov in cov_results}
    
    # 3. Otimização por dia - com cache de resultados intermediários
    print("Iniciando otimização de pesos...")
    
    def optimize_date(date):
        w0_full = df_weights.loc[date].values
        Sigma = cov_dict[date]

        # Dimensão reduzida: ativos com peso != 0 + IBOV + SMAL
        # Usamos máscaras booleanas em vez de where para melhor performance
        active_mask = np.abs(w0_full) > eps
        active_mask[ibov_idx] = True
        active_mask[smal_idx] = True
        active = np.where(active_mask)[0]
        
        w0 = w0_full[active]
        Sigma_sub = Sigma[np.ix_(active, active)]

        # Índices de long/short no subvetor - usando máscaras
        long_mask = w0 > eps
        short_mask = w0 < -eps
        long_idx = np.where(long_mask)[0]
        short_idx = np.where(short_mask)[0]
        
        # Localiza índices de IBOV e SMAL no subvetor
        idx_ibov_sub = int(np.where(active == ibov_idx)[0][0])
        idx_smal_sub = int(np.where(active == smal_idx)[0][0])
        orig_ibov_smal = w0[idx_ibov_sub] + w0[idx_smal_sub]

        # Bounds de ajuste otimizados - agora calculados como percentual da posição original
        bounds = []
        for j, orig in enumerate(w0):
            if j in (idx_ibov_sub, idx_smal_sub):
                bounds.append((orig, 0.0))
            else:
                if orig > eps:  # long
                    # Calcula o limite como percentual da posição original
                    delta = abs(orig) * peso_limite_pct
                    bounds.append((max(0.0, orig - delta), orig + delta))
                elif orig < -eps:  # short
                    # Calcula o limite como percentual da posição original
                    delta = abs(orig) * peso_limite_pct
                    bounds.append((orig - delta, min(0.0, orig + delta)))
                else:  # zero
                    bounds.append((0.0, 0.0))

        # Restrições lineares
        def constraint_long(x):
            return np.sum(x[long_idx]) - 1.0
            
        def constraint_short(x):
            return np.sum(x[short_idx]) + 1.0
            
        def constraint_ibov_smal(x):
            return x[idx_ibov_sub] + x[idx_smal_sub] - (orig_ibov_smal + 0.05)
        
        cons = [
            {'type': 'eq', 'fun': constraint_long},
            {'type': 'eq', 'fun': constraint_short},
            {'type': 'ineq', 'fun': constraint_ibov_smal}
        ]

        # Objetivo com penalização de volatilidade - mais eficiente
        def obj(x):
            diff = x - w0
            norm_term = np.sum(diff * diff)  # Mais eficiente que produto interno
            # Calcula var_term utilizando multiplicação de matriz otimizada
            var_term = np.dot(x, np.dot(Sigma_sub, x)) * 252
            penalty = (var_term - target_vol**2)**2
            return norm_term + lambda_pen * penalty

        # Otimização com menos iterações para casos mais fáceis
        res = minimize(
            obj, w0, bounds=bounds, constraints=cons,
            options={'maxiter': 300, 'ftol': 1e-5, 'eps': 1e-5}
        )

        w_opt = res.x
        
        # Constrói o vetor completo de pesos de forma eficiente
        w_full = np.zeros(n_assets)
        w_full[active] = w_opt
        w_full[np.abs(w_full) < eps] = 0.0

        # Calcula volatilidade final
        vol_real = np.sqrt(np.dot(w_opt, np.dot(Sigma_sub, w_opt)) * 252)
        
        return date, w_full, vol_real, res.success

    # 4. Execução em paralelo com barra de progresso
    results = Parallel(n_jobs=n_jobs, backend='loky')(
        delayed(optimize_date)(d) for d in tqdm(dates, desc="Otimizando pesos")
    )

    # 5. Monta outputs de forma mais eficiente
    print("Finalizando e organizando resultados...")
    
    # Separamos os resultados para melhor eficiência de memória
    dates_list = [r[0] for r in results]
    weights_list = [r[1] for r in results]
    vols_list = [r[2] for r in results]
    success_list = [r[3] for r in results]
    
    # Criamos os DataFrames e Series de forma mais eficiente
    df_weights_adj = pd.DataFrame(
        weights_list,
        index=dates_list,
        columns=assets
    )
    
    series_vol_seeking = pd.Series(
        vols_list,
        index=dates_list,
        name='vol_seeking'
    )
    
    series_converged = pd.Series(
        success_list,
        index=dates_list,
        name='converged'
    )
    
    # Para garantir consistência com o output original - corrigindo os warnings de depreciação
    df_weights_adj = df_weights_adj.reindex(all_dates).ffill().fillna(0.0)
    series_vol_seeking = series_vol_seeking.reindex(all_dates).ffill()
    
    # Para tratar o warning sobre downcasting de objetos booleanos
    series_converged_reindexed = series_converged.reindex(all_dates)
    series_converged_reindexed = series_converged_reindexed.fillna(False)
    # Explicitamente garantir que o tipo é booleano
    series_converged = series_converged_reindexed.astype(bool)
    
    # Mostrar tempo total de execução
    tempo_total = time.time() - tempo_inicio
    print(f"Otimização concluída em {tempo_total:.2f} segundos")
    
    return df_weights_adj, series_vol_seeking, series_converged

# Exemplo de chamada da função com as datas de otimização
# Carregando o arquivo de datas
try:
    df_datas_otimizacao = pd.read_excel("datas_otimizacao.xlsx")
    print(f"Carregadas {len(df_datas_otimizacao)} datas de otimização do arquivo.")
except Exception as e:
    print(f"Erro ao carregar arquivo de datas de otimização: {e}")
    df_datas_otimizacao = None

# Chamada da função com as datas específicas
[df_weights_seeking, series_vol_seeking, series_converged] = volatility_seeking_parallel(
    df_precos_longos, df_posicoes_net, df_posicoes_ibov, df_posicoes_smal, 
    datas_otimizacao=df_datas_otimizacao
)

# Salvar resultados
df_weights_seeking.to_excel("pesos_otimizados.xlsx")
pd.DataFrame({
    'vol_seeking': series_vol_seeking,
    'converged': series_converged
}).to_excel("metricas_otimizacao.xlsx")

print("Resultados salvos em pesos_otimizados.xlsx e metricas_otimizacao.xlsx")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

# Definir tema personalizado para gráficos profissionais
pio.templates["custom_theme"] = pio.templates["plotly_white"].update(
    layout=dict(
        font=dict(family="Arial, sans-serif", size=12),
        title=dict(font=dict(size=20, color="#333")),
        legend=dict(
            title=dict(font=dict(size=13)),
            font=dict(size=12),
            bgcolor="rgba(255, 255, 255, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.1)",
            borderwidth=1
        ),
        margin=dict(t=120, b=80, l=80, r=80),  # Margens amplas
        plot_bgcolor="white",
        paper_bgcolor="white",
        colorway=["#3366CC", "#DC3912", "#FF9900", "#109618", "#990099", "#0099C6"]
    )
)

def visualizar_volatilidade(series_vol_seeking, vol_expost, target_vol=0.06):
    """
    Cria uma visualização interativa comparando volatilidade otimizada vs original
    com design aprimorado e elementos informativos adicionais.
    """
    # Preparar dados
    df = pd.concat([series_vol_seeking, vol_expost], axis=1)
    df.columns = ['Volatilidade Otimizada', 'Volatilidade Original']
    
    # Calcular diferença e melhoria percentual
    df['Redução de Volatilidade'] = df['Volatilidade Original'] - df['Volatilidade Otimizada']
    df['% Melhoria'] = (df['Redução de Volatilidade'] / df['Volatilidade Original']) * 100
    
    # Calcular estatísticas para o subtítulo
    reducao_media = df['Redução de Volatilidade'].mean()
    melhoria_media = df['% Melhoria'].mean()
    max_melhoria = df['% Melhoria'].max()
    
    # Criar gráfico de base com subplots
    fig = make_subplots(
        rows=2, cols=1, 
        shared_xaxes=True,
        vertical_spacing=0.22,  # Espaçamento amplo entre os gráficos
        row_heights=[0.7, 0.3],
        subplot_titles=(
            None,  # Sem título no primeiro subplot
            "Redução de Volatilidade"
        )
    )
    
    # Adicionar linhas de volatilidade no primeiro subplot
    fig.add_trace(
        go.Scatter(
            x=df.index,
            y=df['Volatilidade Otimizada'],
            name="Volatilidade Otimizada",
            line=dict(color="#3366CC", width=2.5),
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Volatilidade: %{y:.2%}<extra></extra>"
        ),
        row=1, col=1
    )
    
    fig.add_trace(
        go.Scatter(
            x=df.index,
            y=df['Volatilidade Original'],
            name="Volatilidade Original",
            line=dict(color="#DC3912", width=2.5, dash='dot'),
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Volatilidade: %{y:.2%}<extra></extra>"
        ),
        row=1, col=1
    )
    
    # Adicionar linha do target
    fig.add_shape(
        type="line",
        x0=df.index.min(),
        x1=df.index.max(),
        y0=target_vol,
        y1=target_vol,
        line=dict(color="#109618", width=2, dash="dash"),
        row=1, col=1
    )
    
    # Adicionar anotação do target
    fig.add_annotation(
        x=df.index.max(),
        y=target_vol,
        text=f"Target: {target_vol:.1%}",
        showarrow=False,
        font=dict(color="#109618", size=12, family="Arial, sans-serif"),
        xanchor="right",
        yanchor="bottom",
        xshift=-5,
        yshift=5,
        row=1, col=1
    )
    
    # Adicionar gráfico de barras da redução de volatilidade
    fig.add_trace(
        go.Bar(
            x=df.index,
            y=df['Redução de Volatilidade'],
            name="Redução de Volatilidade",
            marker_color="rgba(51, 102, 204, 0.6)",  # Azul semi-transparente
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Redução: %{y:.2%}<br>Melhoria: %{customdata:.2f}%<extra></extra>",
            customdata=df['% Melhoria'].values
        ),
        row=2, col=1
    )
    
    # Adicionar média móvel da redução
    ma_window = 21  # média móvel de 21 dias (aproximadamente 1 mês)
    df['MA_Reducao'] = df['Redução de Volatilidade'].rolling(window=ma_window).mean()
    
    fig.add_trace(
        go.Scatter(
            x=df.index,
            y=df['MA_Reducao'],
            name=f"Média Móvel ({ma_window}d)",
            line=dict(color="#990099", width=2.5),
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Média Móvel: %{y:.2%}<extra></extra>"
        ),
        row=2, col=1
    )
    
    # Adicionar linha zero para referência
    fig.add_shape(
        type="line",
        x0=df.index.min(),
        x1=df.index.max(),
        y0=0,
        y1=0,
        line=dict(color="gray", width=1),
        row=2, col=1
    )
    
    # Título principal e subtítulo com estatísticas
    titulo_principal = "Comparação de Volatilidade Ex-Post Anualizada"
    subtitulo = f"Redução média: {reducao_media:.2%} | Melhoria média: {melhoria_media:.2f}% | Melhoria máxima: {max_melhoria:.2f}%"
    
    # Configurar layout
    fig.update_layout(
        template="custom_theme",
        title=dict(
            text=f"{titulo_principal}<br><span style='font-size:14px; color:#666666'>{subtitulo}</span>",
            x=0.5,
            xanchor="center",
            y=0.97,
            yanchor="top"
        ),
        legend=dict(
            orientation="h",
            yanchor="top",
            y=0.94,  # Posicionado abaixo do título
            xanchor="right",
            x=0.99,
            bgcolor="rgba(255, 255, 255, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.1)",
            borderwidth=1
        ),
        hovermode="x unified",
        xaxis=dict(
            rangeselector=dict(
                buttons=[
                    dict(count=1, label="1m", step="month", stepmode="backward"),
                    dict(count=3, label="3m", step="month", stepmode="backward"),
                    dict(count=6, label="6m", step="month", stepmode="backward"),
                    dict(count=1, label="1a", step="year", stepmode="backward"),
                    dict(step="all", label="Tudo")
                ],
                bgcolor="rgba(255, 255, 255, 0.8)",
                bordercolor="rgba(0, 0, 0, 0.1)",
                borderwidth=1,
                x=0.01,
                y=1.115,
            ),
            rangeslider=dict(visible=True, thickness=0.05),
            type="date"
        ),
        xaxis2=dict(
            type="date",
            rangeslider=dict(visible=False)
        ),
        yaxis=dict(
            title=dict(
                text="Volatilidade Anualizada",
                font=dict(size=14)
            ),
            tickformat=".1%",
            gridcolor="rgba(0, 0, 0, 0.1)",
            zerolinecolor="rgba(0, 0, 0, 0.3)"
        ),
        yaxis2=dict(
            title=dict(
                text="Redução de Volatilidade",
                font=dict(size=14)
            ),
            tickformat=".1%",
            gridcolor="rgba(0, 0, 0, 0.1)",
            zerolinecolor="rgba(0, 0, 0, 0.3)"
        ),
        height=800,
        width=1200
    )
    
    return fig, df

def visualizar_alteracoes_pesos(df_weights_original, df_weights_seeking, eps=1e-6):
    """
    Cria uma visualização interativa das alterações nos pesos do portfólio
    com design aprimorado e métricas adicionais.
    """
    # Garantir alinhamento de dados
    common_index = df_weights_original.index.intersection(df_weights_seeking.index)
    common_columns = df_weights_original.columns.intersection(df_weights_seeking.columns)
    
    df_orig = df_weights_original.loc[common_index, common_columns]
    df_seek = df_weights_seeking.loc[common_index, common_columns]
    
    # Calcular alterações
    df_diff = df_seek - df_orig
    df_diff_abs = df_diff.abs()
    
    # Preparar dados para o gráfico
    result_data = []
    
    for date in common_index:
        orig_weights = df_orig.loc[date]
        diffs = df_diff.loc[date]
        diffs_abs = df_diff_abs.loc[date]
        
        # Identificar posições long e short
        long_mask = orig_weights > eps
        short_mask = orig_weights < -eps
        
        # Filtrar colunas especiais
        exclude_cols = ['IBOV', 'SMAL11']
        for col in exclude_cols:
            if col in common_columns:
                long_mask.loc[col] = False
                short_mask.loc[col] = False
        
        # Calcular métricas usando seleção direta de índices
        long_indices = long_mask[long_mask].index
        short_indices = short_mask[short_mask].index
        
        long_changes_sum = diffs_abs.loc[long_indices].sum() if len(long_indices) > 0 else 0
        short_changes_sum = diffs_abs.loc[short_indices].sum() if len(short_indices) > 0 else 0
        total_changes_sum = long_changes_sum + short_changes_sum
        
        # Contar alterações significativas (>0.1%)
        sig_changes = (diffs_abs > 0.001).sum()
        
        # Calcular turnover como soma de alterações / 2
        turnover = total_changes_sum / 2
        
        result_data.append({
            'Data': date,
            'Alterações Long': long_changes_sum,
            'Alterações Short': short_changes_sum,
            'Alterações Totais': total_changes_sum,
            'Turnover': turnover,
            'N° Alterações Significativas': sig_changes
        })
    
    # Criar dataframe com os resultados
    df_changes = pd.DataFrame(result_data).set_index('Data')
    
    # Calcular estatísticas para o subtítulo
    turnover_medio = df_changes['Turnover'].mean()
    alteracoes_media = df_changes['N° Alterações Significativas'].mean()
    
    # Criar gráfico base com 3 subplots claramente separados
    fig = make_subplots(
        rows=3, cols=1,
        shared_xaxes=True,
        vertical_spacing=0.12,
        row_heights=[0.4, 0.3, 0.3],
        subplot_titles=(
            "Magnitude das Alterações por Tipo de Posição",
            "Turnover do Portfólio",
            "Número de Alterações Significativas"
        )
    )
    
    # PRIMEIRO PAINEL: Alterações Long e Short como áreas empilhadas
    # Primeiro Long (verde)
    fig.add_trace(
        go.Scatter(
            x=df_changes.index,
            y=df_changes['Alterações Long'],
            name="Alterações Long",
            fill='tozeroy',
            line=dict(color="#109618", width=0),  # Linha invisível
            fillcolor="rgba(16, 150, 24, 0.6)",   # Verde semi-transparente
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Alterações Long: %{y:.2%}<extra></extra>"
        ),
        row=1, col=1
    )
    
    # Depois Short (vermelho) - empilhado sobre Long
    fig.add_trace(
        go.Scatter(
            x=df_changes.index,
            y=df_changes['Alterações Short'],
            name="Alterações Short",
            fill='tonexty',  # Preencher até o traço anterior
            line=dict(color="#DC3912", width=0),  # Linha invisível
            fillcolor="rgba(220, 57, 18, 0.6)",   # Vermelho semi-transparente
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Alterações Short: %{y:.2%}<extra></extra>"
        ),
        row=1, col=1
    )
    
    # SEGUNDO PAINEL: Turnover como barras
    fig.add_trace(
        go.Bar(
            x=df_changes.index,
            y=df_changes['Turnover'],
            name="Turnover",
            marker_color="rgba(255, 153, 0, 0.6)",  # Laranja semi-transparente
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Turnover: %{y:.2%}<extra></extra>"
        ),
        row=2, col=1
    )
    
    # Adicionar média móvel do turnover
    ma_window = 21
    df_changes['MA_Turnover'] = df_changes['Turnover'].rolling(window=ma_window).mean()
    
    fig.add_trace(
        go.Scatter(
            x=df_changes.index,
            y=df_changes['MA_Turnover'],
            name=f"Média Móvel ({ma_window}d)",
            line=dict(color="#990099", width=2.5),
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Média Móvel: %{y:.2%}<extra></extra>"
        ),
        row=2, col=1
    )
    
    # TERCEIRO PAINEL: Número de alterações significativas
    fig.add_trace(
        go.Bar(
            x=df_changes.index,
            y=df_changes['N° Alterações Significativas'],
            name="N° Alterações Significativas",
            marker_color="rgba(0, 153, 198, 0.6)",  # Azul turquesa semi-transparente
            hovertemplate="Data: %{x|%d/%m/%Y}<br>Alterações: %{y}<extra></extra>"
        ),
        row=3, col=1
    )
    
    # Título principal e subtítulo com estatísticas
    titulo_principal = "Alterações nos Pesos do Portfólio"
    subtitulo = f"Turnover médio: {turnover_medio:.2%} | N° médio de alterações: {alteracoes_media:.1f}"
    
    # Configurar layout
    fig.update_layout(
        template="custom_theme",
        title=dict(
            text=f"{titulo_principal}<br><span style='font-size:14px; color:#666666'>{subtitulo}</span>",
            x=0.5,
            xanchor="center",
            y=0.98,
            yanchor="top"
        ),
        legend=dict(
            orientation="h",
            yanchor="top",
            y=0.95,  # Posicionado abaixo do título
            xanchor="right",
            x=0.99,
            bgcolor="rgba(255, 255, 255, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.1)",
            borderwidth=1
        ),
        hovermode="x unified",
        xaxis=dict(
            rangeselector=dict(
                buttons=[
                    dict(count=1, label="1m", step="month", stepmode="backward"),
                    dict(count=3, label="3m", step="month", stepmode="backward"),
                    dict(count=6, label="6m", step="month", stepmode="backward"),
                    dict(count=1, label="1a", step="year", stepmode="backward"),
                    dict(step="all", label="Tudo")
                ],
                bgcolor="rgba(255, 255, 255, 0.8)",
                bordercolor="rgba(0, 0, 0, 0.1)",
                borderwidth=1,
                x=0.01,
                y=1.08
            ),
            rangeslider=dict(visible=True, thickness=0.05),
            type="date"
        ),
        xaxis2=dict(
            type="date",
            rangeslider=dict(visible=False)
        ),
        xaxis3=dict(
            type="date",
            rangeslider=dict(visible=False)
        ),
        yaxis=dict(
            title=dict(
                text="Magnitude das Alterações",
                font=dict(size=14)
            ),
            tickformat=".1%",
            gridcolor="rgba(0, 0, 0, 0.1)",
            zerolinecolor="rgba(0, 0, 0, 0.3)"
        ),
        yaxis2=dict(
            title=dict(
                text="Turnover",
                font=dict(size=14)
            ),
            tickformat=".1%",
            gridcolor="rgba(0, 0, 0, 0.1)",
            zerolinecolor="rgba(0, 0, 0, 0.3)"
        ),
        yaxis3=dict(
            title=dict(
                text="N° de Alterações",
                font=dict(size=14)
            ),
            gridcolor="rgba(0, 0, 0, 0.1)",
            zerolinecolor="rgba(0, 0, 0, 0.3)"
        ),
        height=900,
        width=1200
    )
    
    return fig, df_changes

def gerar_visualizacoes(series_vol_seeking, vol_expost, df_weights_original, df_weights_seeking):
    """
    Função principal para gerar as visualizações.
    """
    # Gerar visualizações individuais
    fig_vol, df_vol_data = visualizar_volatilidade(series_vol_seeking, vol_expost)
    fig_pesos, df_changes = visualizar_alteracoes_pesos(df_weights_original, df_weights_seeking)
    
    # Exibir visualizações
    fig_vol.show()
    fig_pesos.show()
    
    return {
        'fig_vol': fig_vol,
        'fig_pesos': fig_pesos,
        'df_vol_data': df_vol_data,
        'df_changes': df_changes
    }

# Exemplo de execução
resultados = gerar_visualizacoes(series_vol_seeking, vol_expost, df_posicoes_net, df_weights_seeking)
```
